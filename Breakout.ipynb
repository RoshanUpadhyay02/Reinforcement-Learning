```python
import gym
import numpy as np

env = gym.make('Breakout-v0')
ss = env.observation_space.shape[0]
a_s = env.action_space.n

Q = np.zeros((ss, a_s))

alpha = 0.1
y = 0.99
E = 1.0
E_decay = 0.995
min_E = 0.01
ep = 100
ms = 500

for e in range(ep):
    s = env.reset()
    G = 0
    for step in range(ms):
        if np.random.uniform() < E:
            a = env.action_space.sample()
        else:
            a = np.argmax(Q[s, :])
            
        s1, r, done, info = env.step(a)

        Q[s, a] = (1 - alpha) * Q[s, a] + alpha * (r + y * np.max(Q[s1, :]))

        s = s1
        G += r

        if done:
            break

    E = max(min_E, E * E_decay)

    print("Episode: {}, total_reward: {:.2f}".format(e, G))
```
