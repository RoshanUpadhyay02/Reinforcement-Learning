```python
import numpy as np
import matplotlib.pyplot as plt

n = 10

Es = [0, 0.1, 0.01]

rs = np.zeros(n)

avg_rewards = np.zeros((len(Es), n))

ep = 1000

for e in range(ep):
    s = np.random.randint(0, n)

    for i, E in enumerate(Es):
        if np.random.random() < E:
            a = np.random.randint(0, n)
        else:
            a = np.argmax(rs)

        r = np.random.normal(0, 1)
        
        rs[a] += r
        
        avg_rewards[i][a] += r / (e + 1)

for i, E in enumerate(Es):
    plt.plot(range(n), avg_rewards[i], label='Epsilon = {}'.format(E))

plt.xlabel("Arm")
plt.ylabel("Average Reward")
plt.legend()
plt.title("Epsilon-Greedy Bandit")
plt.show()
```
